```python
# prompt: generate a line plot comparing "TURB" and "LAG" over time "T" label the axes "Time" and "Normalized Fluorescence" show the points and do not include a grid.


import matplotlib.pyplot as plt


#plt.plot(T, TURB, marker='o', label='Turbidostat', color = 'k')
#plt.plot(T, LAG, marker='o', label='Lagoon', color = 'b')
plt.plot(T, TURB, color = 'orange')
plt.plot(T, LAG, color = 'steelblue')

plt.rcParams['font.family'] = 'sans-serif'  # Change font family
plt.rcParams['font.size'] = 9        # Change font size



plt.xlabel("Time (min)")
plt.ylabel("Normalized GFP (RFU)")

plt.grid(False)
plt.savefig('Timecourse.png')
plt.show()
```
