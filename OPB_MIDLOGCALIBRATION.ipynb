```python
## SETUP CELL ##
import numpy as np
import scipy.stats as st
from scipy.stats import ttest_ind
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## IMPORT DATA FILES ##
TCOURSE = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQwQ6kU9JbbKh7kHV6yv4ZCdgDXGm0LOaj6GX9G23OFap_8pJWmrS6gMm687Pkch5aGMGoJKBWBMzQ0/pub?output=csv')


TMAINTAIN = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTXoLWQZm44V5cpnJMESSu2syVmXkqP9JjSDozZ7f-DpLolLbv4n4955nwVIAPH8g/pub?output=csv')
```

```python
## IN TCOURSE replace the column "Time" with a range from 0-37526 ##

TCOURSE['Time (hours)'] = range(0, 37526)
```

```python

COURSE = pd.DataFrame({
    'Time (hours)': range(37526),  # Create the 'Time' column with values from 0 to 74900
    'Relative Optical Density': TCOURSE['10s Average'], # Assuming '10s Average' column exists in TCOURSE
    'Maintain': TMAINTAIN['10s Average'].iloc[:37526] # Assuming '10s Average' column exists in TMAINTAIN
})

```

```python
## copy the plot above, but label the tick marks with  ['1h', '2h', '3h', '4h', '5h', '6h', '7h', '8h'] ##

sns.lineplot(x="Time (hours)", y="Relative Optical Density", data=COURSE.iloc[:30000], color='black')
sns.lineplot(x="Time (hours)", y="Maintain", data=COURSE.iloc[:30000], color ='steelblue')


# Add a horizontal line at y=475
plt.axhline(y=475, color='red', linestyle='--')

# Set x-axis ticks and labels
plt.xticks(np.arange(0, 30001, 3600), ['0','1', '2', '3', '4', '5', '6', '7', '8'])

plt.rcParams['font.family'] = 'sans-serif'  # Change font family
plt.rcParams['font.size'] = 9     # Change font size

plt.show()
```

```python

# Assuming TCOURSE DataFrame is already loaded as in the previous code

plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization

sns.lineplot(x="Time (hr)", y="Measure", data=TCOURSE.iloc[:30000], color='steelblue', label='Measure') # Set alpha for transparency
sns.lineplot(x="Time (hr)", y="10s Average", data=TCOURSE.iloc[:30000], color='skyblue', label='10s Average')

# Set x-axis ticks and labels
plt.xticks(np.arange(0, 30001, 3600), ['0h','1h', '2h', '3h', '4h', '5h', '6h', '7h', '8h'])


plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Overlayed Line Plots")
plt.legend()
plt.savefig('Tcourse.svg', format='svg')
plt.grid(True) # Add grid for better readability
plt.show()
```
